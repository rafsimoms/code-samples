## Setup

```python
from movielens_analysis import Ratings, Movies, Tags, Links

%timeit ratings = Ratings('../datasets/ratings.csv')
ratings = Ratings('../datasets/ratings.csv')
%timeit movies_obj = Movies('../datasets/movies.csv')
movies_obj = Movies('../datasets/movies.csv')
%timeit tags_obj = Tags('../datasets/tags.csv')
tags_obj = Tags('../datasets/tags.csv')
%timeit links_obj = Links('../datasets/links.csv')
links_obj = Links('../datasets/links.csv')

ratings_movies = ratings.get_movies()
ratings_users = ratings.get_users()

print('All classes initialized successfully.')
```

---
## 1. Ratings — Movies subclass

### 1.1 `Ratings.Movies.dist_by_year()` — Distribution of ratings by year

```python
%timeit ratings_movies.dist_by_year()

result = ratings_movies.dist_by_year()
print('Ratings count by year (ascending):')
for year, count in result.items():
    print(f'  {year}: {count}')
```

### 1.2 `Ratings.Movies.dist_by_rating()` — Distribution of ratings by rating value

```python
%timeit ratings_movies.dist_by_rating()

result = ratings_movies.dist_by_rating()
print('Ratings count by rating value (ascending):')
for rating, count in result.items():
    print(f'  {rating}: {count}')
```

### 1.3 `Ratings.Movies.top_by_num_of_ratings(n)` — Top-10 movies by number of ratings

```python
%timeit ratings_movies.top_by_num_of_ratings(10)

result = ratings_movies.top_by_num_of_ratings(10)
print('Top-10 movies by number of ratings:')
for title, count in result.items():
    print(f'  {title}: {count}')
```

### 1.4 `Ratings.Movies.top_by_ratings(n, metric)` — Top-10 movies by average rating

```python
%timeit ratings_movies.top_by_ratings(10, 'average')

result = ratings_movies.top_by_ratings(10, 'average')
print('Top-10 movies by average rating:')
for title, avg in result.items():
    print(f'  {title}: {avg}')
```

### 1.5 `Ratings.Movies.top_by_ratings(n, metric)` — Top-10 movies by median rating

```python
%timeit ratings_movies.top_by_ratings(10, 'median')

result = ratings_movies.top_by_ratings(10, 'median')
print('Top-10 movies by median rating:')
for title, med in result.items():
    print(f'  {title}: {med}')
```

### 1.6 `Ratings.Movies.top_controversial(n)` — Top-10 most controversial movies (by rating variance)

```python
%timeit ratings_movies.top_controversial(10)

result = ratings_movies.top_controversial(10)
print('Top-10 most controversial movies (highest rating variance):')
for title, var in result.items():
    print(f'  {title}: {var}')
```

```python
%timeit ratings_users.dist_by_ratings('median')

result = ratings_users.dist_by_ratings('median')
top10 = list(result.items())[:10]
print('Top-10 users by median rating given:')
for user_id, med in top10:
    print(f'  User {user_id}: {med}')
```

## 3. Movies class

### 3.1 'dist_by_release()' - метод анализирует датасет и возвращает количество фильмов снятых в каждый из годов.

```python

%timeit -n 1 -r 1 movies_obj.dist_by_release()

result = movies_obj.dist_by_release()
print('Movies count by release year (sorted by count):')
for year, count in result.items():
    print(f'  {year}: {count}')
```

### 3.2 dist_by_genres() - метод анализирует датасет и возвращает количество фильмов снятых в каждом из жанров

```python
%timeit movies_obj.dist_by_genres()

result = movies_obj.dist_by_genres()
print('Movies count by genres:')
for genre, count in result.items():
    print(f'  {genre}: {count}')
```

### most_genres(n) - метод анализирует датасет и возвращает топ n  фильмов с наибольшим количеством жанров.

```python
%timeit movies_obj.most_genres(10)

result = movies_obj.most_genres(10)
print('Top movies by number of genres:')
for movie, num in result.items():
    print(f'  {movie}: {num}')
```

## 4. Tags class

### 4.1 most_words(n) -  возвращает топ n тегов с наибольшим количеством слов

```python
%timeit tags_obj.most_words(10)

result = tags_obj.most_words(10)

print("Tags with the largest number of words:")
for tag, word_count in result.items():
    print(f'"{tag}" — {word_count} words')
```

### 4.2 longest(n) - возвращает топ n тегов с наибольшим количеством символов

```python
%timeit tags_obj.longest(10)

result = tags_obj.longest(10)

print("Longest tags:")
for tag, length in result:
    print(f'"{tag}" — {length} characters')
```

### 4.3 most_words_and_longest(n) - находит пересечения среди топ n самых длинных по словам и по символам тегов 

```python
%timeit tags_obj.most_words_and_longest(10)

result = tags_obj.most_words_and_longest(10)

print("Tags that are both longest and contain the most words:")
for tag in result:
    print(tag)
```

### 4.4 most_popular(n) - находит n самых популярных тегов 

```python
%timeit tags_obj.most_popular(10)

result = tags_obj.most_popular(10)

print("Most popular tags:")
for tag, count in result.items():
    print(f'"{tag}" — {count}')
```

### 4.5 tags_with(word) - находит все теги, содержащие заданное слово word.

```python
%timeit tags_obj.tags_with("movie")

result = tags_obj.tags_with("movie")

print('Tags containing the word "movie":')
for tag in result:
    print(tag)
```

## 5. Links class

### 5.1 `Links.get_imdb(list_of_movies, list_of_fields)` — Fetch IMDB data for specific movies

```python
sample_movie_ids = [links_obj.data[i][0] for i in range(3)]
fields = ['Director', 'Budget', 'Gross', 'Runtime']

%timeit -n 1 -r 1 links_obj.get_imdb(sample_movie_ids, fields)

result = links_obj.get_imdb(sample_movie_ids, fields)
print(f'IMDB data for movie IDs {sample_movie_ids}:')
for row in result:
    print(f'  movieId={row[0]} | Director={row[1]} | Budget={row[2]} | Gross={row[3]} | Runtime={row[4]}')
```

### 5.2 `Links.top_directors(n)` — Top-5 directors by number of movies

```python
%timeit -n 1 -r 1 links_obj.top_directors(5)

result = links_obj.top_directors(5)
print('Top-5 directors by number of movies in dataset:')
for director, count in result.items():
    print(f'  {director}: {count} movies')
```

### 5.3 `Links.most_expensive(n)` — Top-5 most expensive movies by budget

```python
%timeit -n 1 -r 1 links_obj.most_expensive(5)

result = links_obj.most_expensive(5)
print('Top-5 most expensive movies by budget:')
for title, budget in result.items():
    print(f'  {title}: {budget}')
```

### 5.4 `Links.most_profitable(n)` — Top-5 most profitable movies (gross − budget)

```python
%timeit -n 1 -r 1 links_obj.most_profitable(5)

result = links_obj.most_profitable(5)
print('Top-5 most profitable movies (gross - budget):')
for title, profit in result.items():
    print(f'  {title}: ${profit:,}')
```

### 5.5 `Links.longest(n)` — Top-5 longest movies by runtime

```python
%timeit -n 1 -r 1 links_obj.longest(5)

result = links_obj.longest(5)
print('Top-5 longest movies by runtime:')
for title, minutes in result.items():
    print(f'  {title}: {minutes} min')
```

### 5.6 `Links.top_cost_per_minute(n)` — Top-5 movies by budget per minute of runtime

```python
%timeit -n 1 -r 1 links_obj.top_cost_per_minute(5)

result = links_obj.top_cost_per_minute(5)
print('Top-5 movies by cost per minute of runtime:')
for title, cost in result.items():
    print(f'  {title}: ${cost:,.2f} per minute')
```
