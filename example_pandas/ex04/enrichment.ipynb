```python
import pandas as pd
import numpy as np

df = pd.read_json('../data/auto.json')
df
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
sample_df = df.sample(n=200, replace=True, random_state=21)

sample_df['Fines'] = np.random.choice(df['Fines'], size=200)
sample_df['Refund'] = np.random.choice(df['Refund'], size=200)
```

```python
concat_rows = pd.concat([df, sample_df], ignore_index=True)
```

```python
np.random.seed(21)
years = pd.Series(np.random.randint(1980, 2020, size=len(concat_rows)), name='Year')

fines = pd.concat([concat_rows.reset_index(drop=True), years], axis=1)
```

```python
surnames = pd.read_json('../../datasets/surname.json')

surnames.columns = surnames.iloc[0] 
surnames = surnames[1:].reset_index(drop=True)
import re
unique_car_numbers = fines['CarNumber'].unique()
np.random.seed(21)
random_surnames = np.random.choice(
    [re.sub(r'\W', '', s) for s in surnames['NAME']],
    size=len(unique_car_numbers)
)
owners = pd.DataFrame({
    'CarNumber': unique_car_numbers,
    'SURNAME': random_surnames
})
```

```python
owners.head()
```

```python
new_fines = pd.DataFrame({
    'CarNumber': ['NEW001', 'NEW002', 'NEW003', 'NEW004', 'NEW005'],
    'Make': ['Ford', 'Toyota', 'BMW', 'Audi', 'Kia'],
    'Model': ['Focus', 'Corolla', 'X5', 'A4', 'Sportage'],
    'Refund': [1, 2, 1, 2, 1],
    'Fines': [1000, 2000, 1500, 2500, 1800],
    'Year': [2001, 2005, 2010, 2015, 2018]
})

fines = pd.concat([fines, new_fines], ignore_index=True)

owners = owners[:-20]
owners_extra = pd.DataFrame({
    'CarNumber': ['OWN001', 'OWN002', 'OWN003'],
    'SURNAME': ['Smith', 'Johnson', 'Brown']
})
owners = pd.concat([owners, owners_extra], ignore_index=True)
```

```python

inner_join = pd.merge(fines, owners, on='CarNumber', how='inner')

outer_join = pd.merge(fines, owners, on='CarNumber', how='outer')

left_join = pd.merge(fines, owners, on='CarNumber', how='left')

right_join = pd.merge(fines, owners, on='CarNumber', how='right')
```
